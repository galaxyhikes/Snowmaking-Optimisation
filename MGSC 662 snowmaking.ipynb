```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load slope status matrices (assumed to be defined as beginner_trail_matrix, intermediate_trail_matrix, and hard_trail_matrix)

# Parameters
num_rows, num_cols = 100, 200  # Grid dimensions

# Temperature parameters
top_temp = -5.5
bottom_temp = -3.5
temperature_gradient = np.linspace(top_temp, bottom_temp, num_rows).reshape(-1, 1)
temperature_matrix = np.tile(temperature_gradient, (1, num_cols))

# Initialize the combined matrix with three binary dimensions for slope levels and one for temperature
combined_matrix = np.zeros((num_rows, num_cols, 5), dtype=int)

# Populate combined matrix based on slope levels, enforcing a priority of hard > intermediate > beginner
for i in range(num_rows):
    for j in range(num_cols):
        if hard_trail_matrix[i, j]:  # Priority for hardest level
            combined_matrix[i, j, 0] = 0  # Beginner
            combined_matrix[i, j, 1] = 0  # Intermediate
            combined_matrix[i, j, 2] = 1  # Hard
            combined_matrix[i, j, 4] = True  # Slope exists
        elif intermediate_trail_matrix[i, j]:  # Next priority
            combined_matrix[i, j, 0] = 0  # Beginner
            combined_matrix[i, j, 1] = 1  # Intermediate
            combined_matrix[i, j, 2] = 0  # Hard
            combined_matrix[i, j, 4] = True  # Slope exists
        elif beginner_trail_matrix[i, j]:  # Lowest priority
            combined_matrix[i, j, 0] = 1  # Beginner
            combined_matrix[i, j, 1] = 0  # Intermediate
            combined_matrix[i, j, 2] = 0  # Hard
            combined_matrix[i, j, 4] = True  # Slope exists

# Assign temperature values to the 4th dimension for each cell where there's a slope
for i in range(num_rows):
    for j in range(num_cols):
        if combined_matrix[i, j, 0] == 1 or combined_matrix[i, j, 1] == 1 or combined_matrix[i, j, 2] == 1:
            combined_matrix[i, j, 3] = round(temperature_matrix[i, j], 1)

# Display output for a specific cell
print("Slope Level (Beginner, Intermediate, Hard) and Temperature at [68, 60]:", combined_matrix[68, 60])
print("Beginner:", combined_matrix[68, 60, 0])  # 1 if Beginner, else 0
print("Intermediate:", combined_matrix[68, 60, 1])  # 1 if Intermediate, else 0
print("Hard:", combined_matrix[68, 60, 2])  # 1 if Hard, else 0
print("Temperature:", combined_matrix[68, 60, 3])  # Temperature value if a slope exists at this point
print("Slope Exists:", combined_matrix[68, 60, 4])  # True if there is a slope, else False
```

```python
import gurobipy as gp
from gurobipy import GRB
import numpy as np

height, width = 100, 200
coverage_size = 7  # Size of coverage (7x7)
min_temp = -2.5
min_distance = 3

snow_matrix = combined_matrix[..., 4].astype(int)

model = gp.Model("snowgun_placement")
model.setParam('MIPGap', 0.03)

#Snowgun placement (0 and 1 matrix)
x = model.addVars(height, width, vtype=GRB.BINARY, name="x")

# Objective is to minimize the number of snowguns
model.setObjective(gp.quicksum(x[i, j] for i in range(height) for j in range(width)), GRB.MINIMIZE)

# Be sure that every snow area is covered by a snowgun
for i in range(height):
    for j in range(width):
        if snow_matrix[i, j] == 1: 
            coverage_indices = []
            for di in range(-coverage_size // 2, coverage_size // 2 + 1):
                for dj in range(-coverage_size // 2, coverage_size // 2 + 1):
                    ni, nj = i + di, j + dj
                    if 0 <= ni < height and 0 <= nj < width:
                        coverage_indices.append(x[ni, nj])
            model.addConstr(gp.quicksum(coverage_indices) >= 1, name=f"cover_{i}_{j}")

# Temperature constraints
for i in range(height):
    for j in range(width):
        model.addConstr(
            x[i, j] * (temperature_matrix[i, j] - min_temp) <= 0,
                name=f"temp_constraint_{i}_{j}"
                )

# Minimum distance apart constraint
for di in range(-min_distance, min_distance + 1):
    for dj in range(-min_distance, min_distance + 1):
        ni, nj = i + di, j + dj
        if (di != 0 or dj != 0) and (0 <= ni < height) and (0 <= nj < width) and (i < ni or (i == ni and j < nj)):
            model.addConstr(x[i, j] + x[ni, nj] <= 1, name=f"min_dist_{i}_{j}_{ni}_{nj}")

for i in range(height):
    for j in range(width):
        # Define the region within the minimum distance
        for di in range(-min_distance, min_distance + 1):
            for dj in range(-min_distance, min_distance + 1):
                ni, nj = i + di, j + dj
                # Skip the current cell (i, j) and ensure the neighbor is within bounds
                if (di != 0 or dj != 0) and (0 <= ni < height) and (0 <= nj < width) and (i < ni or (i == ni and j < nj)):
                    # If x[i, j] = 1, then x[ni, nj] must be 0
                    model.addConstr(x[i, j] + x[ni, nj] <= 1, name=f"min_dist_{i}_{j}_{ni}_{nj}")

model.optimize()

placement_matrix = np.zeros((height, width), dtype=int)
if model.status == GRB.OPTIMAL:
    for i in range(height):
        for j in range(width):
            placement_matrix[i, j] = int(x[i, j].x)

# placement matrix
print("Snow gun placement matrix:")
print(placement_matrix)
```

```python
height, width = 100, 200
coverage_size = [8, 6, 7, 3, 3]
snowguns = ['Fan gun Titan', 'Fan gun Evo', 'Fan gun Ventus', 'Lance EOS', 'Lance DUO']
power_consumption = [23, 14.2, 20, 2, 4.8]
water_consumption = [43200, 28800, 32400, 24840, 28800]
min_temp = [-2.5,-2.5,-2.5,-4,-4]
min_distance =[3, 3, 3, 2, 2]

# Costs(electricity cost = 12.5 cents/kWh, water cost = $3.16/m)
electricity_cost = [pc * 0.125 for pc in power_consumption]
water_cost = [wc * 3.16 / 1000 for wc in water_consumption]
cost_per_snowgun = [e + w for e, w in zip(electricity_cost, water_cost)]

snow_matrix = combined_matrix[..., 4].astype(int)

model = gp.Model("snowgun_placement")
model.setParam('MIPGap', 0.005)

x = model.addVars(height, width, len(snowguns), vtype=GRB.BINARY, name="x")

# Objective function: minimize total cost
model.setObjective(
    gp.quicksum(x[i, j, k] * cost_per_snowgun[k] for i in range(height) for j in range(width) for k in range(len(snowguns))),
    GRB.MINIMIZE
)

# Constraint : every cell must be covered
for i in range(height):
    for j in range(width):
        if snow_matrix[i, j] == 1:  # Cell requires snow coverage
            coverage_indices = []
            for k in range(len(snowguns)):  # For each snow gun type
                for di in range(-coverage_size[k] // 2, coverage_size[k] // 2 + 1):
                    for dj in range(-coverage_size[k] // 2, coverage_size[k] // 2 + 1):
                        ni, nj = i + di, j + dj
                        if 0 <= ni < height and 0 <= nj < width:
                            coverage_indices.append(x[ni, nj, k])
            model.addConstr(gp.quicksum(coverage_indices) >= 1, name=f"cover_{i}_{j}")

# Placement constraints: At most one snow gun can be placed in a single cell
for i in range(height):
    for j in range(width):
        model.addConstr(gp.quicksum(x[i, j, k] for k in range(len(snowguns))) <= 1, name=f"place_{i}_{j}")

# Temperature constraint
for i in range(height):
    for j in range(width):
        for k in range(len(snowguns)):
            model.addConstr(
                x[i, j, k] * (temperature_matrix[i, j] - min_temp[k]) <= 0,
                name=f"temp_constraint_{i}{j}{k}"
            )

# Minimum distance constraint
for i in range(height):
    for j in range(width):
        for k1 in range(len(snowguns)):  # Loop for first snowgun type
            for k2 in range(len(snowguns)):  # Loop for second snowgun type (can be same or different)
                for di in range(-min_distance[max(k1, k2)], min_distance[max(k1, k2)] + 1):  # Checking for neighbors within the distance
                    for dj in range(-min_distance[max(k1, k2)], min_distance[max(k1, k2)] + 1):
                        ni, nj = i + di, j + dj
                        # Skip the current cell (i, j) and ensure the neighbor is within bounds
                        if (di != 0 or dj != 0) and 0 <= ni < height and 0 <= nj < width:
                            # If a snowgun of type k1 is placed at (i, j) and a snowgun of type k2 is placed at (ni, nj),
                            # they must be at least min_distance apart
                            model.addConstr(x[i, j, k1] + x[ni, nj, k2] <= 1, 
                                             name=f"min_dist_{i}_{j}_{ni}_{nj}_{k1}_{k2}")


model.optimize()

placement_array = np.zeros((height, width, len(snowguns)), dtype=int)
if model.status == GRB.OPTIMAL:
    for i in range(height):
        for j in range(width):
            for k in range(len(snowguns)):
                placement_array[i, j, k] = int(x[i, j, k].x)

print("Snow gun placement matrix:")
print(placement_array)

snowgun_counts = {snowgun: 0 for snowgun in snowguns}

for k, snowgun in enumerate(snowguns):
    snowgun_counts[snowgun] = sum(int(x[i, j, k].x) for i in range(height) for j in range(width))

print("Number of each type of snow gun placed:")
for snowgun, count in snowgun_counts.items():
    print(f"{snowgun}: {count}")
```

```python
# dimensions
height, width = 100, 200
coverage_size = [8, 6, 7, 3, 3]
total_snow_area = np.sum(combined_matrix[..., 4])  # Total number of cells requiring snow
snowguns = ['Fan gun Titan', 'Fan gun Evo', 'Fan gun Ventus', 'Lance EOS', 'Lance DUO']
power_consumption = [23, 14.2, 20, 2, 4.8]
water_consumption = [43200, 28800, 32400, 24840, 28800]
snow_production = [120, 80, 90, 69, 80]  # Snow production rates (m/h) for each square meter 
min_temp = [-2.5,-2.5,-2.5,-4,-4]
min_distance =[3, 3, 3, 2, 2]

# Costs (electricity cost = 12.5 cents/kWh, water cost = $3.16/m)
electricity_cost = [pc * 0.125 for pc in power_consumption]
water_cost = [wc * 3.16 / 1000 for wc in water_consumption]
cost_per_snowgun = [e + w for e, w in zip(electricity_cost, water_cost)]

# Coverage threshold as a percentage (e.g., 90%)
coverage_threshold = 1

# Minimum snow production requirements (m/h) for each slope type
min_snow_per_cell = {
    "Beginner": 500,  # Minimum m/h required per cell
    "Intermediate": 300,
    "Hard": 200,
}

# Extract slope levels from the combined matrix
beginner_mask = combined_matrix[..., 0]
intermediate_mask = combined_matrix[..., 1]
hard_mask = combined_matrix[..., 2]

model = gp.Model("snowgun_placement")
model.setParam('MIPGap', 0.045)#Limit for not running for too long

# Decision variables: x[i, j, k] (binary, 1 if snowgun of type k is placed at (i, j))
x = model.addVars(height, width, len(snowguns), vtype=GRB.BINARY, name="x")

# Objective function: Minimize total cost
model.setObjective(
    gp.quicksum(x[i, j, k] * cost_per_snowgun[k] for i in range(height) for j in range(width) for k in range(len(snowguns))),
    GRB.MINIMIZE
)

# Define "covered" cells as binary variables
covered = model.addVars(height, width, vtype=GRB.BINARY, name="covered")

# Compute snow production per cell and add constraints
for i in range(height):
    for j in range(width):
        if combined_matrix[i, j, 4] == 1:  # Only consider slope cells
            # Calculate the snow production in this cell from all covering snowguns
            cell_snow_production = gp.quicksum(
                x[ni, nj, k] * snow_production[k]
                for k in range(len(snowguns))
                for ni in range(max(0, i - coverage_size[k] // 2), min(height, i + coverage_size[k] // 2 + 1))
                for nj in range(max(0, j - coverage_size[k] // 2), min(width, j + coverage_size[k] // 2 + 1))
            )
            
            # Determine required snow production for this cell based on its slope type
            required_snow = (
                min_snow_per_cell["Beginner"] * beginner_mask[i, j] +
                min_snow_per_cell["Intermediate"] * intermediate_mask[i, j] +
                min_snow_per_cell["Hard"] * hard_mask[i, j]
            )
            
            # Add constraint: snow production in this cell must meet or exceed the requirement
            model.addConstr(cell_snow_production >= required_snow, name=f"snow_production_cell_{i}_{j}")

# Ensure the total covered area meets the threshold
model.addConstr(
    gp.quicksum(covered[i, j] for i in range(height) for j in range(width)) >= coverage_threshold * total_snow_area,
    name="coverage_threshold"
)

# Placement constraints: At most one snowgun can be placed in a single cell
for i in range(height):
    for j in range(width):
        model.addConstr(gp.quicksum(x[i, j, k] for k in range(len(snowguns))) <= 1, name=f"place_{i}_{j}")

# Temperature constraint (based on slope type)
for i in range(height):
    for j in range(width):
        for k in range(len(snowguns)):
            # Check if the cell is on a specific slope type and apply the temperature constraint accordingly
            if beginner_mask[i, j] == 1:  # Beginner slope
                model.addConstr(
                    x[i, j, k] * (temperature_matrix[i, j] - min_temp[k]) <= 0,
                    name=f"temp_constraint_{i}_{j}_{k}_beginner"
                )
            elif intermediate_mask[i, j] == 1:  # Intermediate slope
                model.addConstr(
                    x[i, j, k] * (temperature_matrix[i, j] - min_temp[k]) <= 0,
                    name=f"temp_constraint_{i}_{j}_{k}_intermediate"
                )
            elif hard_mask[i, j] == 1:  # Hard slope
                model.addConstr(
                    x[i, j, k] * (temperature_matrix[i, j] - min_temp[k]) <= 0,
                    name=f"temp_constraint_{i}_{j}_{k}_hard"
                )

# Minimum distance constraint (based on slope type)
for k1 in range(len(snowguns)):
    for k2 in range(k1, len(snowguns)):  # Check pairs of snowgun types (including same type)
        min_dist = max(min_distance[k1], min_distance[k2])  # Max of the two distances

        # Apply constraints based on positions and slopes
        for i in range(height):
            for j in range(width):
                for ni in range(height):
                    for nj in range(width):
                        if abs(i - ni) + abs(j - nj) <= min_dist:  # Check Manhattan distance condition
                            if (i != ni or j != nj):  # Don't apply to the same cell
                                # Apply constraint differently based on slope type
                                if beginner_mask[i, j] == 1 and beginner_mask[ni, nj] == 1:  # Both are beginner slopes
                                    model.addConstr(
                                        x[i, j, k1] + x[ni, nj, k2] <= 1,
                                        name=f"distance_constraint_beginner_{i}_{j}_{ni}_{nj}_{k1}_{k2}"
                                    )
                                elif intermediate_mask[i, j] == 1 and intermediate_mask[ni, nj] == 1:  # Both are intermediate slopes
                                    model.addConstr(
                                        x[i, j, k1] + x[ni, nj, k2] <= 1,
                                        name=f"distance_constraint_intermediate_{i}_{j}_{ni}_{nj}_{k1}_{k2}"
                                    )
                                elif hard_mask[i, j] == 1 and hard_mask[ni, nj] == 1:  # Both are hard slopes
                                    model.addConstr(
                                        x[i, j, k1] + x[ni, nj, k2] <= 1,
                                        name=f"distance_constraint_hard_{i}_{j}_{ni}_{nj}_{k1}_{k2}"
                                    )

model.optimize()

placement_array = np.zeros((height, width, len(snowguns)), dtype=int)
if model.status == GRB.OPTIMAL:
    for i in range(height):
        for j in range(width):
            for k in range(len(snowguns)):
                placement_array[i, j, k] = int(x[i, j, k].x)

print("Snow gun placement matrix:")
print(placement_array)

# Print the number of each type of snowgun placed
snowgun_counts = {snowgun: 0 for snowgun in snowguns}

for k in range(len(snowguns)):
        snowgun_counts[snowguns[k]] = sum(int(x[i, j, k].x) for i in range(height) for j in range(width))

print("Number of each type of snowgun placed:")
for snowgun, count in snowgun_counts.items():
    print(f"{snowgun}: {count}")
```

```python
from gurobipy import Model, GRB, quicksum
import numpy as np

# Initialize the model
model = Model("snowgun_cost_security")
model.setParam('MIPGap', 0.1)

# Dimensions and data
height, width = 100, 200
coverage_size = [8, 6, 7, 3, 3]
snowguns = ['Fan gun Titan', 'Fan gun Evo', 'Fan gun Ventus', 'Lance EOS', 'Lance DUO']
power_consumption = [23, 14.2, 20, 2, 4.8]
water_consumption = [43200, 28800, 32400, 24840, 28800]
snow_production = [120, 80, 90, 69, 80]
min_snow_per_cell = {"Beginner": 400, "Intermediate": 200, "Hard": 100}
min_temp = [-2.5,-2.5,-2.5,-4,-4]
min_distance =[3, 3, 3, 2, 2]

# Costs (electricity cost = 12.5 cents/kWh, water cost = $3.16/m)
electricity_cost = [pc * 0.125 for pc in power_consumption]
water_cost = [wc * 3.16 / 1000 for wc in water_consumption]
cost_per_snowgun = [e + w for e, w in zip(electricity_cost, water_cost)]

# Combined matrix
black_cells = combined_matrix[..., 4]  # Binary matrix indicating black cells (slopes)

# Slope levels
beginner_mask = combined_matrix[..., 0]
intermediate_mask = combined_matrix[..., 1]
hard_mask = combined_matrix[..., 2]

# Total snow area to cover
total_snow_area = np.sum(combined_matrix[..., 4])  # Total number of cells requiring snow

# Decision variables: x[i, j, k] (binary, 1 if snowgun of type k is placed at (i, j))
x = model.addVars(height, width, len(snowguns), vtype=GRB.BINARY, name="x")

# Compute total cost
total_cost = quicksum(
    x[i, j, k] * cost_per_snowgun[k] for i in range(height) for j in range(width) for k in range(len(snowguns))
)

# Compute security score
security_score = quicksum(
    x[i, j, k] * (
        3 * black_cells[i, j] +
        2 * (
            black_cells[max(0, i - 1), j] +
            black_cells[min(height - 1, i + 1), j] +
            black_cells[i, max(0, j - 1)] +
            black_cells[i, min(width - 1, j + 1)]
        ) +
        1 * (
            black_cells[max(0, i - 2), j] +
            black_cells[min(height - 1, i + 2), j] +
            black_cells[i, max(0, j - 2)] +
            black_cells[i, min(width - 1, j + 2)]
        )
    )
    for i in range(height) for j in range(width) for k in range(len(snowguns))
)

# Objective weights
weight_cost = 1  # Weight for cost (to minimize)
weight_security = 1  # Weight for security (to maximize; negative because we minimize overall)

# Set the combined objective
model.setObjective(weight_cost * total_cost - weight_security * security_score, GRB.MINIMIZE)

# Add constraints for snow production
for i in range(height):
    for j in range(width):
        if combined_matrix[i, j, 4] == 1:  # Only consider slope cells
            # Snow production from all covering snowguns
            cell_snow_production = quicksum(
                x[ni, nj, k] * snow_production[k]
                for k in range(len(snowguns))
                for ni in range(max(0, i - coverage_size[k] // 2), min(height, i + coverage_size[k] // 2 + 1))
                for nj in range(max(0, j - coverage_size[k] // 2), min(width, j + coverage_size[k] // 2 + 1))
            )
            
            # Required snow production for this cell
            required_snow = (
                min_snow_per_cell["Beginner"] * beginner_mask[i, j] +
                min_snow_per_cell["Intermediate"] * intermediate_mask[i, j] +
                min_snow_per_cell["Hard"] * hard_mask[i, j]
            )
            
            # Add snow production constraint
            model.addConstr(cell_snow_production >= required_snow, name=f"snow_production_cell_{i}_{j}")

# Ensure the total covered area meets the threshold
coverage_threshold = 1
covered = model.addVars(height, width, vtype=GRB.BINARY, name="covered")

model.addConstr(
    quicksum(covered[i, j] for i in range(height) for j in range(width)) >= coverage_threshold * total_snow_area,
    name="coverage_threshold"
)

# Placement constraints: At most one snowgun can be placed in a single cell
for i in range(height):
    for j in range(width):
        model.addConstr(quicksum(x[i, j, k] for k in range(len(snowguns))) <= 1, name=f"place_{i}_{j}")

# Temperature constraint (based on slope type)
for i in range(height):
    for j in range(width):
        for k in range(len(snowguns)):
            # Check if the cell is on a specific slope type and apply the temperature constraint accordingly
            if beginner_mask[i, j] == 1:  # Beginner slope
                model.addConstr(
                    x[i, j, k] * (temperature_matrix[i, j] - min_temp[k]) <= 0,
                    name=f"temp_constraint_{i}_{j}_{k}_beginner"
                )
            elif intermediate_mask[i, j] == 1:  # Intermediate slope
                model.addConstr(
                    x[i, j, k] * (temperature_matrix[i, j] - min_temp[k]) <= 0,
                    name=f"temp_constraint_{i}_{j}_{k}_intermediate"
                )
            elif hard_mask[i, j] == 1:  # Hard slope
                model.addConstr(
                    x[i, j, k] * (temperature_matrix[i, j] - min_temp[k]) <= 0,
                    name=f"temp_constraint_{i}_{j}_{k}_hard"
                )


# Distance constraint
for k1 in range(len(snowguns)):
    for k2 in range(len(snowguns)):
        min_dist = max(min_distance[k1], min_distance[k2])  # Max distance for the pair
        for i in range(height):
            for j in range(width):
                for ni in range(height):
                    for nj in range(width):
                        if abs(i - ni) + abs(j - nj) <= min_dist and (i != ni or j != nj):
                            model.addConstr(
                                x[i, j, k1] + x[ni, nj, k2] <= 1,
                                name=f"min_dist_constraint_{i}_{j}_{ni}_{nj}_{k1}_{k2}"
                            )

# Solve the model
model.optimize()

# Extract the solution
placement_array = np.zeros((height, width, len(snowguns)), dtype=int)
if model.status == GRB.OPTIMAL:
    for i in range(height):
        for j in range(width):
            for k in range(len(snowguns)):
                placement_array[i, j, k] = int(x[i, j, k].x)

# Display the result
print("Snow gun placement matrix:")
print(placement_array)
```
